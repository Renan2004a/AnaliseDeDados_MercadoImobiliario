```python
#leitor de csv
import pandas as pd

dados = pd.read_csv('/content/drive/MyDrive/melb_data.csv')

num_linhas = 10

pd.set_option('display.max_columns', None)  # Para exibir as colunas
pd.set_option('display.max_rows', num_linhas)  # Para exibir as linhas

traducoes = {
    'Suburb': 'Bairro',
    'Address': 'Endereço',
    'Rooms': 'Quartos',
    'Type': 'Tipo',
    'Price': 'Preço',
    'Method': 'Método',
    'SellerG': 'Vendedor',
    'Date': 'Data',
    'Distance': 'Distância',
    'Postcode': 'CEP',
    'Bedroom2': 'Quartos_secundários',
    'Bathroom': 'Banheiro',
    'Car': 'Vagas_de_garagem',
    'Landsize': 'Tamanho_do_terreno',
    'BuildingArea': 'Área_construída',
    'YearBuilt': 'Ano_de_construção',
    'CouncilArea': 'Área_do_conselho',
    'Lattitude': 'Latitude',
    'Longtitude': 'Longitude',
    'Regionname': 'Nome_da_região',
    'Propertycount': 'Contagem_de_propriedades'
}

dados = dados.rename(columns=traducoes)
tabela = pd.DataFrame(dados)
display(tabela)
```

```python
# Histograma
import matplotlib.pyplot as plt
import pandas as pd

# Seleciona as colunas numéricas do DataFrame 'dados'
dados_numericos = dados.select_dtypes(include=['float', 'int'])

# Visualizações para variáveis numéricas
for coluna in dados_numericos.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(dados[coluna], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histograma de {coluna}')
    plt.xlabel('Valor')
    plt.ylabel('Frequência')
    plt.grid(True)
    plt.show()
```

```python
# Calcular a matriz de correlação

dados_numericos = dados.select_dtypes(include=['number'])

tabela_descritiva_com_correlacao = dados_numericos.corr()


pd.set_option('display.max_rows', None)

print("Tabela de Correlação:")
display(tabela_descritiva_com_correlacao)
```

```python
# Identificar as correlações mais fortes

abs_matriz_correlacao = tabela_descritiva_com_correlacao.abs()

melhores_correlacoes = abs_matriz_correlacao.unstack().sort_values(ascending=False)
melhores_correlacoes = melhores_correlacoes[melhores_correlacoes != 1]

print("Top 10 Melhores Correlações:")
print(melhores_correlacoes.head(10))
```

```python
#Qual é a relação entre o número de quartos e o preço de aluguel?
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(dados['Quartos'], dados['Preço'], alpha=0.5)
plt.title('Relação entre Número de Quartos e Preço de Aluguel')
plt.xlabel('Número de Quartos')
plt.ylabel('Preço de Aluguel')
plt.grid(True)
plt.show()

correlacao_quartos_preco = dados['Quartos'].corr(dados['Preço'])
print("Correlação entre o número de quartos e o preço de aluguel:", correlacao_quartos_preco)
```

Correlação Positiva Moderada: à medida que o número de quartos de um imóvel aumenta, o preço de aluguel tende a aumentar também, mas não de forma proporcional.

```python
# Os imóveis mais antigos tendem a ter preços de aluguel mais baixos?
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(dados['Ano_de_construção'], dados['Preço'], alpha=0.5)
plt.title('Relação entre Idade dos Imóveis e Preço de Aluguel')
plt.xlabel('Ano de Construção')
plt.ylabel('Preço de Aluguel')
plt.grid(True)
plt.show()

correlacao_idade_preco = dados['Ano_de_construção'].corr(dados['Preço'])
print("Correlação entre a idade dos imóveis e o preço de aluguel:", correlacao_idade_preco)
```

```python
# Média dos preços de aluguel para cada bairro
pd.set_option('display.max_rows', None)
media_precos_por_bairro = dados.groupby('Bairro')['Preço'].mean().round(2)

media_precos_por_bairro_ordenado = media_precos_por_bairro.sort_values()

print("Média dos Preços de Aluguel por Bairro (do mais barato para o mais caro):")
print(media_precos_por_bairro_ordenado)
```

```python
#Como os preços de aluguel variaram ao longo dos anos em diferentes bairros?
import matplotlib.pyplot as plt

cinco_bairros_principais = dados.groupby('Bairro')['Preço'].mean().sort_values(ascending=False).head(5).index

dados_cinco_bairros_principais = dados[dados['Bairro'].isin(cinco_bairros_principais)]

media_precos_por_ano_bairro = dados_cinco_bairros_principais.groupby(['Ano_de_construção', 'Bairro'])['Preço'].mean().unstack()


plt.figure(figsize=(12, 8))

for bairro in media_precos_por_ano_bairro.columns:
    plt.bar(media_precos_por_ano_bairro.index, media_precos_por_ano_bairro[bairro], label=bairro)

plt.title('Variação dos Preços de Aluguel ao Longo dos Anos nos Cinco Bairros Principais')
plt.xlabel('Ano')
plt.ylabel('Preço Médio de Aluguel')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calcular o teste Qui-Quadrado
from scipy.stats import chi2_contingency

# Selecionar apenas as variáveis categóricas
dados_categoricos = dados.select_dtypes(include=['object'])

for var1 in dados_categoricos.columns:
    for var2 in dados_categoricos.columns:
        if var1 != var2:
            contingency_table = pd.crosstab(dados[var1], dados[var2])
            chi2, p, _, _ = chi2_contingency(contingency_table)
            print(f"Qui-Quadrado entre '{var1}' e '{var2}': p-valor = {p:.4f}")
```

```python
import random

pd.set_option('display.max_rows', None)
bairros_unicos = dados['Bairro'].unique()

# Escolha aleatória de 10 bairros
principais_bairros = random.sample(list(bairros_unicos), 10)

# Tabela cruzada de Bairro e Método para os bairros escolhidos aleatoriamente
cross_tab_bairro_metodo_principais = pd.crosstab(dados[dados['Bairro'].isin(principais_bairros)]['Bairro'], dados[dados['Bairro'].isin(principais_bairros)]['Método'])

print("\nTabela Cruzada de Bairro e Método para 10 Bairros Aleatórios:")
display(cross_tab_bairro_metodo_principais)

# Plotar gráfico de barras empilhadas (stacked bar chart)
cross_tab_bairro_metodo_principais.plot(kind='bar', stacked=True, figsize=(12, 8))


plt.title('Tabela Cruzada de Bairro e Método para 10 Bairros Aleatórios')
plt.xlabel('Bairro')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.legend(title='Método')

plt.show()

```
